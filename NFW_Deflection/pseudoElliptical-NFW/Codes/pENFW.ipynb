# pseudo Elliptical NFW model

Here we implement, as a test script, the pseudo elliptical NFW model for lensing. 

The full references are:
1. [Pseudo Elliptical Models]    - https://www.aanda.org/articles/aa/pdf/2012/08/aa18485-11.pdf

For pseudo ellipitcal lens models it is useful to introduce the elliptical coordinates

\begin{equation}
x_1 = \frac{x_\epsilon}{\sqrt{a_1}}\cos{\phi_\epsilon}, \,\, x_2 = \frac{x_\epsilon}{\sqrt{a_2}}\sin{\phi_\epsilon},
\end{equation}

where $(x_1, x_2)$ are cartesian coordinate, $x_\epsilon = \sqrt{a_1 x_1^2 + a_2 x_2^2}$ and $\phi_\epsilon = \arctan{ \left(x_2/x_1 \sqrt{a_2/a_1} \right)}$. Beyond that, $a_1$ and $a_2$, are respectively the semi major and semi minor axis of an ellipse, defining the ellipticity by $\epsilon = 1 - \sqrt{a_2/a_1}$.
