```python
import numpy as np
from astropy.io import fits
from matplotlib import rc
from matplotlib.lines import Line2D

from shapely.geometry import Point
from shapely.geometry.polygon import Polygon
```

```python
#Reading mask data (shoud be in ascii file with two columns -- ra dec)

x_arc_l,y_arc_l=np.loadtxt('maior.txt',unpack=True)
x_arc_s,y_arc_s=np.loadtxt('menor.txt',unpack=True)


#Reading image
image_data = fits.getdata('Alma_with_lens_center.fits')

image_shape = image_data.shape

mask = np.ones(image_shape)

```

```python
print(image_shape)
```

```python

pol_l=Polygon(zip(x_arc_l,y_arc_l))
pol_s=Polygon(zip(x_arc_s,y_arc_s))



for i in range(image_shape[1]):
    for j in range(image_shape[0]):
        point = Point((i,j))
        if pol_l.contains(point):
            mask[j,i]=0
        if pol_s.contains(point):
            mask[j,i]=0       
            
fits.writeto('mask.fits',mask,clobber=True)            
```
