Obj: After making the cut in the original image of the HST and determining the center by the MGE, this code centers the central pixel of the MGE with the middle of the image. Usualy before the rotation.

```python
import numpy as np
import matplotlib.pyplot as plt
import astropy.units as u
from mpdaf.obj import Image, WCS
import astropy.units as u
from mpdaf.obj import deg2sexa, sexa2deg
data_path 
```

```python
F160W = Image('/home/carlos/Desktop/HST SDP81/Data/Corte_F160w.fits') #Open the image file F475w
F110W.info() #Display information about the file
F160W.info() #Display information about the file
```

```python
CentralPx = (0.651817, 135.798) #Central pixel. The format is (y,x)
```

```python
F110W_centerized = F110W.subimage(center=CentralPx, unit_center=u.deg, size=15., unit_size=u.arcsec)
F160W_centerized = F160W.subimage(center=CentralPx, unit_center=u.deg, size=15., unit_size=u.arcsec)
    #center: determine the new center
    #unit_center: astropy units of center coordenate. None for px units
    #size: if you disere a new cut, size define the width
    #unit_size: astropy units for the size. None for px units
```

```python
F110W_centerized.info()
F160W_centerized.info()
```

```python
folder_save = '/home/carlos/Desktop/HST SDP81/Data/' #Path for the save folder
F110W_centerized.write(folder_save + 'F110W_centerized.fits') #Save the new centralized images
F160W_centerized.write(folder_save + 'F160W_centerized.fits')
```
