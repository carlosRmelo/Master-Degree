```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Bin, vel, disp, chi2, dV, dsigma = np.loadtxt('Output_Ppxf_BadAssTemplate.txt', unpack=True) #Reads the table with the adjustment data/LÃª a tabela com os dados do ajuste

vel = vel - vel[0]                                   #vel[0] velocity of central px
vrms = np.sqrt(np.array(vel)**2 + np.array(disp)**2) #Vrms velocity
erms = np.sqrt((np.array(dV)*np.array(vel))**2 + (np.array(dsigma)*np.array(disp))**2)/vrms #error in vrms
```

```python
xnode = np.array([ 0.        ,  0.        , -0.2       , -0.3       , -0.5       ,
       -0.32      ,  0.13333333, -1.02      , -0.8       , -1.22857143,
       -0.3       , -0.33333333, -0.55151515,  0.1       , -0.05      ,
        0.1       ,  0.2       ,  0.2       ,  0.4       ,  0.32      ,
        0.53333333,  0.5       ,  0.18285714,  1.03478261,  0.8       ,
        1.03333333,  1.48275862,  0.57647059])

ynode = np.array([ 0.        , -0.3       , -0.3       ,  0.        , -0.3       ,
       -0.96      , -0.66666667, -0.86      ,  0.05714286,  0.45714286,
        0.2       ,  0.46666667,  1.38787879,  0.2       ,  0.75      ,
        0.4       ,  0.        , -0.3       , -0.1       ,  0.64      ,
        0.13333333, -0.5       , -1.55428571, -0.9826087 , -0.2       ,
        0.56666667, -0.05517241,  1.34117647])
r = np.sqrt(xnode**2 + ynode**2)
```

```python
plt.figure(figsize=(8,6))
plt.errorbar(r,vrms, yerr=erms, fmt=".k", capsize=0)
plt.ylabel("Vrms (km/s)")
plt.xlabel("Circularized radius")
```
