```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Bin, vel, disp, chi2, dV, dsigma = np.loadtxt('Output_Ppxf_BadAssTemplate.txt', unpack=True) #Reads the table with the adjustment data/LÃª a tabela com os dados do ajuste

vel = vel - vel[0]                                   #vel[0] velocity of central px
vrms = np.sqrt(np.array(vel)**2 + np.array(disp)**2) #Vrms velocity
erms = np.sqrt((np.array(dV)*np.array(vel))**2 + (np.array(dsigma)*np.array(disp))**2)/vrms #error in vrms
```

```python
xnode = np.array([ 0.        , -0.3       , -0.33333333, -0.13333333, -1.15      ,
       -0.97142857, -0.08      , -0.10212766,  0.28      ,  0.2       ,
        0.56      ,  1.14736842,  0.35      ,  0.        ,  1.24571429,
       -0.72352941,  0.43809524])
ynode = np.array([-0.1       , -0.3       ,  0.06666667,  0.26666667, -0.13571429,
        0.77142857,  0.64      ,  1.48085106,  0.44      , -0.1       ,
        0.08      ,  0.69473684, -0.4       , -0.6       , -0.50857143,
       -1.11764706, -1.42380952])
r = np.sqrt(xnode**2 + ynode**2)
```

```python
plt.figure(figsize=(8,6))
plt.errorbar(r,vrms, yerr=erms, fmt=".k", capsize=0)
plt.ylabel("Vrms (km/s)")
plt.xlabel("Circularized radius")
```
