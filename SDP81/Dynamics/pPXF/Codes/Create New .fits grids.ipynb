Obj: Create a new grid where the value of the pixel correspond to the Vorbin Id.

```python
from mpdaf.obj import Image, WCS, Cube
import numpy as np
from astropy import units as u
```

```python
data = Cube('/home/carlos/Desktop/SDP/Dynamics/pPXF/Data/Cut.fits')
data.info()
```

```python
y_px, x_px, Id = np.loadtxt('Voronoi_OutPut.txt', unpack=True) 
```

```python
img = data.sum(axis=0)
img[:] = np.nan
img.var[:] = np.nan
```

```python
for i in range(len(y_px)):                               
    img.data.data[int(y_px[i])][int(x_px[i])] = Id[i]
    img.var[int(y_px[i])][int(x_px[i])] = Id[i]
```

```python
img.write('/home/carlos/Desktop/SDP/Dynamics/pPXF/Data/Grid_with_Vorbin.fits') #Save the new grid
img.info() #Display the information
```
