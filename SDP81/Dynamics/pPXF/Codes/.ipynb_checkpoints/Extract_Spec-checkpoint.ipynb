```python
from mpdaf.obj import Cube
from mpdaf.drs import PixTable
import matplotlib.pyplot as plt
from mpdaf.obj import Image
from mpdaf.obj import iter_spe
import astropy.units as u
from mpdaf.obj import deg2sexa, sexa2deg
import numpy as np
from mpdaf.obj import Spectrum, WaveCoord
```

Opening the cubes.

```python
data = Cube('/home/carlos/Desktop/SDP MUSE (07-12-20)/Data/4800-6300(6arc).fits')
continuum = Cube('/home/carlos/Desktop/SDP MUSE (07-12-20)/Data/Continuum(6arc).fits')
data.info()
continuum.info()
```

Extract complete data.

```python
for sp,(y,x) in iter_spe(data, index=True):
        print("Pixel [%d,%d]" % (y, x))
        sp.write('/home/carlos/Desktop/SDP MUSE (07-12-20)/Spec/%dx%d.fits'%(y,x))
        coord_sky = data.wcs.pix2sky([y,x], unit=u.arcsec)
        y_arcs = coord_sky[0][0]
        x_arcs = coord_sky[0][1] 
        print(y_arcs,x_arcs)
        newrow = [np.int(y), np.int(x), y_arcs, x_arcs]
        if y== 0 and x ==0:
            table = np.array([np.int(y),np.int(x),y_arcs,x_arcs])
        else:
            table = np.vstack([table,newrow])
```

```python
np.savetxt('PixelPosition_data.txt',table, header= 'y_px x_px y_arcs x_arcs', fmt='%i %i %.18e %.18e',delimiter=' ') 
```

Extract continuum.

```python
for spec,(y,x) in iter_spe(continuum, index=True):
        print("Pixel [%d,%d]" % (y, x))
        spec.write('/home/carlos/Desktop/SDP MUSE (07-12-20)/Continuum/%dx%d.fits'%(y,x))
        coord_sky = continuum.wcs.pix2sky([y,x], unit=u.arcsec)
        y_arcs = coord_sky[0][0]
        x_arcs = coord_sky[0][1] 
        print(y_arcs,x_arcs)
        newrow = [np.int(y), np.int(x), y_arcs, x_arcs]
        if y== 0 and x ==0:
            table = np.array([np.int(y),np.int(x),y_arcs,x_arcs])
        else:
            table = np.vstack([table,newrow])
```

```python
np.savetxt('PixelPosition_continuum.txt',table, header= 'y_px x_px y_arcs x_arcs', fmt='%i %i %.18e %.18e',delimiter=' ') 
```
