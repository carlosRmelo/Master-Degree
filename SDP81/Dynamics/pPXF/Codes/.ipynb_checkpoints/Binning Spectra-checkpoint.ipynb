```python
import numpy as np
from mpdaf.obj import Spectrum
import matplotlib.pyplot as plt
from astropy import units as u
from astropy.io import fits
```

```python
vorbin = np.loadtxt('Voronoi_OutPut.txt') #Abre a tabela que contém os spaxels que devem ser binnados

#Para cada bin ele varre toda a tabela acima procurando pelos espectros relacionados à aquele bin. Se encontra
    #um espectro relacionado ao binNum ele soma ao anteriores que possuem o mesmo binNum
    
for binNum in range(29):
    count = 0 #Apenas para ver se está sendo somada a quantidade de espectros correta em cada bin
    dados = Spectrum('/home/carlos/Desktop/SDP MUSE (07-12-20)/Spec/%dx%d.fits'%(0,0),ext=1) #Abre um espectro para que possa ser
    var =  Spectrum('/home/carlos/Desktop/SDP MUSE (07-12-20)/Spec/%dx%d.fits'%(0,0),ext=2)                                                                   #usado na iteração
    dados -= dados #Esse mesmo espectro é subtraído, para que não cause alteração nos demais
    var -= var
    for n in range(len(vorbin)):
        
        if vorbin[n][2] == binNum: #Verifica o binNum
            x = np.int(vorbin[n][1]) #Posição x do espectro que será somado
            y = np.int(vorbin[n][0]) #Posição y do espectro que será somado
            spec = Spectrum('/home/carlos/Desktop/SDP MUSE (07-12-20)/Spec/%dx%d.fits'%(y,x),ext=1) #Abre o espectro que será somado
            noise = Spectrum('/home/carlos/Desktop/SDP MUSE (07-12-20)/Spec/%dx%d.fits'%(y,x),ext=2)
            dados = dados + spec #Soma o atual espectro com os antigos. Como dados == 0 na primeira
            var = var + noise                           #iteração, ele não altera em nada
            count += 1 #Soma 1 se um espectro foi somado
    print('Número do Bin: %d'%binNum) #Informa qual o binNum foi somado
    print('Número de espectros somados: %d'%count) #Informa quantos espectros foram somados
    dados.write('/home/carlos/Desktop/SDP MUSE (07-12-20)/Selected Bin/Bindat%d.fits'%binNum) #Salva o espectro binnado
    var.write('/home/carlos/Desktop/SDP MUSE (07-12-20)/Selected Bin/Binnoise%d.fits'%binNum)
    print('\n ############################################################## \n')
    
```

```python
bin0 = fits.open('/home/carlos/Desktop/SDP MUSE (07-12-20)/Bin/Bindat0.fits')
data = bin0[1].data

bin0_var = fits.open('/home/carlos/Desktop/SDP MUSE (07-12-20)/Bin/Binnoise0.fits')
noise_data = bin0_var[1].data
erro = np.sqrt(noise_data)
```

```python
signal = np.average(data)
noise = np.average(erro)
```
