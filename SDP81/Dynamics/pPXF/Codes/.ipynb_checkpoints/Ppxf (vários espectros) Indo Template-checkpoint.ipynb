```python
import glob
from os import path

from astropy.io import fits
from scipy import ndimage
import numpy as np

import ppxf as ppxf_package
from ppxf.ppxf import ppxf
import ppxf.ppxf_util as util

import matplotlib.pyplot as plt
```

Adapatação do código Ppxf para se aplicar a diversos espectros.
x e y são as posições em px, de modo que servem de contador e também de index para abrir os arquivos de espectro, que foram salvos no formato **y_px X x_px.fits**.

Como discutido no README, os arquivos .fits que contém os espectros possuem três extensões. Note que o pyton, por padrão, começa a contagem do zero.

Deve-se antentar as unidades de comprimento de onda e fluxo. Por padrão, o ppxf trabalha com comprimento de onda em Angstrons pois estamos utilizando a biblioteca estelar do MILES. No nosso caso como vamos normalizar o espectro, as unidades de intensidade não devem gerar problemas.



```python
ppxf_dir = '/home/carlosmelo/Desktop/ppxf'
for y in range(10):
    for x in range (10):
    #LENDO UM ESPECTRO
        file = ('/home/carlosmelo/Documents/Projeto/Testes com o mpdaf/ESO/Espectros/%dX%d.fits'%(y,x)) #Caminho e nome do arquivo do espectro
        hdu = fits.open(file) #Abrindo o arquivo.fits
        gal_lin = hdu[1].data #Lendo os dados
        h1 = hdu[1].header    #Lendo o header dos dados
        
    #LENDO A VARIÂNCIA DO ESPECTRO
        var = hdu[2].data     #Lendo a variância
        h_var = hdu[2].header #Lendo o header da variância
        
    #LOGARITMIZANDO O ESPECTRO E A VARIÂNCIA
        #ESPECTRO
        FWHM_gal = 2.71  #FWHM do telescópio em Angstrons
        lambRange1 = h1['CRVAL1'] + np.array([0.,h1['CDELT1']*(h1['NAXIS1']-1)]) #Está linha gera os comprimentos de onda
                        #pois a variável gal_lin armazena em um array apenas as intensidades, mas não os comprimentos de onda
        galaxy, logLam1, velscale = util.log_rebin(lambRange1, gal_lin) #Logaritimizando o espectro da galáxia
        median_galaxy = np.median(galaxy) #Média do espectro, pois vamos normalizá-lo pela média
        galaxy = galaxy/median_galaxy        #Normalizando o espectro pela média
        lam = np.exp(logLam1)                #Recuperando o comprimento de onda em unidades do SI
        
        #VARIÂNCIA
        lamRange_var = h_var['CRVAL1'] + np.array([0.,h_var['CDELT1']*(h_var['NAXIS1']-1)])
        variance, loglam_var, velscale_var = util.log_rebin(lamRange_var, var)
        noise = np.sqrt(variance)
        noise = noise/median_galaxy         #Normalizando a variância pela mesma média do espectro, para que estejam
                                            #na mesma ordem de grandeza
        
        
    #LENDO UM TEMPLATE PARA SABER O TAMANHO DO VETOR FINAL
        #A ideia é crirar um arraay vazio para que mais tarde seja preenchido com as informações de todos os templetes.
        vazdekis = glob.glob(ppxf_dir + '/miles_models/Mun1.30*.fits') 
        FWHM_tem = 2.51  #FWHM do templete em Angstrons
        velscale_ratio = 2
        hdu = fits.open(vazdekis[0])
        ssp = hdu[0].data
        h2 = hdu[0].header
        
        #LOGARITMIZANDO O TEMPLATE
        lamRange2 = h2['CRVAL1'] + np.array([0., h2['CDELT1']*(h2['NAXIS1'] - 1)])
        sspNew, logLam2, velscale_temp = util.log_rebin(lamRange2, ssp, velscale=velscale/velscale_ratio)
        templates = np.empty((sspNew.size, len(vazdekis)))  #Array com o tamanho necessário
    
    #CALCULANDO A FWHM FINAL PARA REALIZAR A CONVOLUÇÃO DO TEMPLATE E GALÁXIA
        FWHM_dif = np.sqrt(FWHM_gal**2 - FWHM_tem**2)
        sigma = FWHM_dif/2.355/h2['CDELT1'] #Novo sigma da PSF.
    
    #LENDO TODOS OS TEMPLATES E JÁ LOGARITMIZANDO
        for j, file in enumerate(vazdekis):
            hdu = fits.open(file)
            ssp = hdu[0].data
            ssp = ndimage.gaussian_filter1d(ssp, sigma) #Realizando a convolução
            sspNew, logLam2, velscale_temp = util.log_rebin(lamRange2, ssp, velscale=velscale/velscale_ratio)
            templates[:, j] = sspNew/np.median(sspNew)
    
    #ADICIONANDO UM SHIFT PARA COINCIDIR OS COMPRIMENTOS DE ONDA DO ESPECTRO E DO TEMPLATE
            #Nem sempre os comprimentos de onda do templete e do espectro coincidem, por esse motivo é dado um 
        #shift artificial ao templete, mas que após o ajuste é retirado.
        c = 299792.458
        dv = (np.mean(logLam2[:velscale_ratio]) - logLam1[0])*c #Shift
        z = 0.035 #Redsifth aproximado do espectro.
        goodPixels = util.determine_goodpixels(logLam1, lamRange2, z)  #Máscaras
        vel = c*np.log(1 + z)
        start = [vel, 200.] #Chute inicial
    
    #INÍCIO DO AJUSTE
        #Para cada espectro ele realiza o ajuste e printa abaixo o nome do px em que foi realizado o ajuste,
        #bem como as infomações do melhor ajuste (qui², Velocidade, Dispersão, Demais momentos). Então cria uma
        #tabela com as informações necessárias e a salva com o nome 'Output_Ppxf.txt'.
        pp = ppxf(templates, galaxy, noise, velscale, start,
                lam = lam, plot=False, moments=4,
                  degree=6,vsyst=dv, goodpixels=goodPixels, velscale_ratio=velscale_ratio)
        print(y,x)
        print('Vel :%f.\nDisp:%f.'%(pp.sol[0],pp.sol[1]))
    
    #ESCREVENDO UMA TABELA COM OS DADOS DE Y_PX, X_PX,  VELOCIDADE(KM/S) E DISPERSÃO(KM/S)
        newrow = [np.int(y), np.int(x), pp.sol[0], pp.sol[1]]
        if y == 0 and x == 0:
            table = np.array([np.int(y),np.int(x),pp.sol[0], pp.sol[1]])
        else:
            table = np.vstack([table,newrow])
        print('\n####################\n')
```

```python
np.savetxt('Output_Ppxf.txt',table, header= 'y_px x_px vel(km/s) disp(km/s)', fmt='%i %i %.18e %.18e',delimiter=' ') 
```
