Performs a annealing binning in the data, including the ellipticity.

```python
from mpdaf.obj import Image, WCS, Cube
import numpy as np

import matplotlib.pyplot as plt
from plotbin import plot_velfield
from mgefit.find_galaxy import find_galaxy

data_folder = "/home/carlos/Documents/GitHub/Master-Degree/SDP81/Dynamics/New pPXF/"
```

```python
#Frist we read the data cube and transform it in an image
cube = Cube(data_folder+"Data/Cut.fits")
img = cube.sum(axis=0)
img.plot(use_wcs=True,zscale=True, cmap="hot", colorbar="v")
```

```python
#Now we define some properties of the image
#Proporties of the image

sky = img.background()[0]         #counts/px
scale = 0.2                       #arcsec/px
img = img - sky

img.plot(use_wcs=True,zscale=True, cmap="hot", colorbar="v")
```

```python
#Now we find some structural properties using find_my_galaxy
f = find_galaxy(img.data.data, fraction=0.2, plot=1)
```

```python
# To compute the elliptical radius we need that x-axis coincident with the galaxy major axis.
# So we need to rotate the pixel position
def rotate_points(x, y, ang):
    """
    Rotates points conter-clockwise by an angle ANG in degrees.
    Michele cappellari, Paranal, 10 November 2013

    """
    theta = np.radians(ang)
    xNew = x*np.cos(theta) - y*np.sin(theta)
    yNew = x*np.sin(theta) + y*np.cos(theta)

    return xNew, yNew  
```

First, we rotate the position to have the major semi-axis on the x-axis.

Then we calculate the elliptical radius in rings of width dr, with respect to the center of the galaxy [f.ypeak, f.xpeak].

Every pixel within the ring width [r, r + dr] receives the same tag.

Radius greater than that defined by the major semi-axis of the galaxy are masked.



```python
dr = 0.0
tag = 0
while dr < 10:
    for ypx in range(60):
        for xpx in range(60):
            rot_x, rot_y = rotate_points(xpx-f.xpeak,ypx-f.ypeak, 90-f.pa)
            r =  scale * np.sqrt(rot_x**2 + (rot_y/(1-f.eps))**2)
            if dr <= r < dr + 0.2:
                img.data.data[ypx,xpx] =  tag
                if r >= f.majoraxis*scale:
                    img.data.data[ypx,xpx] =  np.nan
    dr = dr + 0.2
    tag = tag + 1

img.var.data[:] = np.nan     #Mask the variance
```

```python
plt.figure(figsize=(10,10))
img.plot(colorbar='v', cmap="hot", use_wcs=True, title="Annealing Binning Tag")

#save in a fits file
img.write(data_folder+"Data/Annealing_binning_tag.fits")
```

```python
#Write a table with y_position, x_position, Bin_Id
x      = []
y      = []
bin_id = []

for ypx in range(img.shape[0]):
    for xpx in range(img.shape[1]):
        if not np.isnan( img.data.data[ypx,xpx] ):
            y.append(ypx)
            x.append(xpx)
            bin_id.append(img.data.data[ypx,xpx])

#Save Table
np.savetxt('Annealing_Bin.txt', np.column_stack([y, x, bin_id]),header="y_px \t x_px \t Bin_Id", fmt=b'%10.6f %10.6f %8i')
```

```python
plt.figure(figsize=(10,8))
plot_velfield.plot_velfield((x-f.xpeak)*scale,(y-f.ypeak)*scale, bin_id, colorbar=True)
plt.title("Anneling Binning Tag")
plt.xlabel("arcsec")
plt.ylabel("arcsec")
```

## Now using only spaxels with SNR > 1.

```python
y_px, x_px, signal, noise = np.loadtxt('GoodPixels.txt', unpack=True)
```

```python
(signal/noise).max()
```

```python
dr = 0.0
tag = 0
img.data.data[:] = np.nan
while dr < 10:
    for i in range(len(y_px)):
        for j in range(len(x_px)):
            rot_x, rot_y = rotate_points(x_px[j]-f.xpeak,y_px[i]-f.ypeak, 90-f.pa)
            r =  scale * np.sqrt(rot_x**2 + (rot_y/(1-f.eps))**2)
            if dr <= r < dr + 0.2:
                img.data.data[int(y_px[i]),int(x_px[j])] =  tag
                
    dr = dr + 0.2
    tag = tag + 1

img.var.data[:] = np.nan     #Mask the variance
```

```python
#Write a table with y_position, x_position, Bin_Id
x      = []
y      = []
bin_id = []

for ypx in range(img.shape[0]):
    for xpx in range(img.shape[1]):
        if not np.isnan( img.data.data[ypx,xpx] ):
            y.append(ypx)
            x.append(xpx)
            bin_id.append(img.data.data[ypx,xpx])

#Save Table
np.savetxt('Annealing_Bin_GoodBins.txt', np.column_stack([y, x, bin_id]),header="y_px \t x_px \t Bin_Id", fmt=b'%10.6f %10.6f %8i')
```

```python
#save in a fits file
img.write(data_folder+"Data/Annealing_binning_Goodbins.fits")

plt.figure(figsize=(10,8))
plot_velfield.plot_velfield((x-f.xpeak)*scale,(y-f.ypeak)*scale, bin_id, colorbar=True)
plt.title("Anneling Binning Tag")
plt.xlabel("arcsec")
plt.ylabel("arcsec")
```
