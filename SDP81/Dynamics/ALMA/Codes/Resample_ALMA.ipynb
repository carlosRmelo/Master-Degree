Goal: Resample ALMA image from 0.005'' to 0.01''.

```python
from mpdaf.obj import Image
from astropy import units as u
```

```python
#Open original image

img = Image("/home/carlos/Desktop/SDP/Dynamics/ALMA/Data/ALMA.fits")
img.info()
img.plot(use_wcs=True, zscale=True,  colorbar='v')
```

```python
#Resample img

resampled_img = img.resample(newdim=420, newstart=None, newstep=0.01, unit_step=u.arcsec, flux=True)
```

```python
resampled_img.info()
resampled_img.plot(use_wcs=True, zscale=True, colorbar='v')
resampled_img.data.data.sum()
```

```python
resampled_img.write("/home/carlos/Desktop/SDP/Dynamics/ALMA/Data/ALMA_resampled.fits")
```
