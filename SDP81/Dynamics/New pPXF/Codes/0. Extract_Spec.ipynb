Obj: Extract the spectra in which pixel for the data cube.This should be done after the cut.

```python
from mpdaf.obj import Cube
from mpdaf.drs import PixTable
import matplotlib.pyplot as plt
from mpdaf.obj import Image
from mpdaf.obj import iter_spe
import astropy.units as u
from mpdaf.obj import deg2sexa, sexa2deg
import numpy as np
from mpdaf.obj import Spectrum, WaveCoord

data_folder = "/home/carlos/Documents/GitHub/Master-Degree/SDP81/Dynamics/New pPXF/"
```

Opening the cubes.

```python
data = Cube(data_folder+'Data/Cut.fits')
data.info()
```

```python
sub = data.select_lambda(lbda_min=4800., lbda_max=5500.0)
sub.info()
```

Extract complete data.

```python
for sp,(y,x) in iter_spe(sub, index=True):
        print("Pixel [%d,%d]" % (y, x))
        sp.write(data_folder+'Data/All Spec/%dx%d.fits'%(y,x))
        coord_sky = data.wcs.pix2sky([y,x], unit=u.arcsec)
        y_arcs = coord_sky[0][0]
        x_arcs = coord_sky[0][1] 
        print(y_arcs,x_arcs)
        newrow = [np.int(y), np.int(x), y_arcs, x_arcs]
        if y== 0 and x ==0:
            table = np.array([np.int(y),np.int(x),y_arcs,x_arcs])
        else:
            table = np.vstack([table,newrow])
```

```python
np.savetxt('PixelPosition_data.txt',table, header= 'y_px x_px y_arcs x_arcs', fmt='%i %i %.18e %.18e',delimiter=' ') 
```
