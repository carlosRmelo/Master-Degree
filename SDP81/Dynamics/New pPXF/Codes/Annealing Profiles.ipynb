```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def component(vel, disp, dV, dsigma):
    vel = vel - vel[0]                                   #vel[0] velocity of central px
    vrms = np.sqrt(np.array(vel)**2 + np.array(disp)**2) #Vrms velocity
    erms = np.sqrt((np.array(dV)*np.array(vel))**2 + (np.array(dsigma)*np.array(disp))**2)/vrms #error in vrms
    
    return vrms, erms
    
```

```python
Bin2, vel2, disp2, chi22, dV2, dsigma2 = np.loadtxt('Annealing_Bin02/Output_Ppxf_IndoTemplate.txt', unpack=True) #Reads the table with the adjustment data/Lê a tabela com os dados do ajuste
vrms2, erms2 = component(vel2, disp2, dV2, dsigma2)
dr = 0
rell2 = np.empty_like(vrms2)
for i in range(len(vrms2)):
    rell2[i] = dr
    dr = dr + 0.2
```

```python
Bin3, vel3, disp3, chi23, dV3, dsigma3 = np.loadtxt('Annealing_Bin03/Output_Ppxf_IndoTemplate.txt', unpack=True) #Reads the table with the adjustment data/Lê a tabela com os dados do ajuste
vrms3, erms3 = component(vel3, disp3, dV3, dsigma3)
dr = 0
rell3 = np.empty_like(vrms3)
for i in range(len(vrms3)):
    rell3[i] = dr
    dr = dr + 0.3
```

```python
Bin4, vel4, disp4, chi24, dV4, dsigma4 = np.loadtxt('Annealing_Bin04/Output_Ppxf_IndoTemplate.txt', unpack=True) #Reads the table with the adjustment data/Lê a tabela com os dados do ajuste
vrms4, erms4 = component(vel4, disp4, dV4, dsigma4)
dr = 0
rell4 = np.empty_like(vrms4)
for i in range(len(vrms4)):
    rell4[i] = dr
    dr = dr + 0.4
```

```python
plt.figure(figsize=(12,6))

plt.errorbar(rell2, vrms2, yerr=erms2, fmt=".r", capsize=0, label="Bin:$0.2''$")
plt.errorbar(rell3, vrms3, yerr=erms3, fmt=".k", capsize=0, label="Bin:$0.3''$")
plt.errorbar(rell4, vrms4, yerr=erms4, fmt=".b", capsize=0, label="Bin:$0.4''$")

plt.ylabel("Vrms (km/s)")
plt.xlabel("Circularized radius")
plt.legend()
```
