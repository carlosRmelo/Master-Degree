Obj: Create a new grid where the value of the pixel correspond to the Vorbin Id.

```python
from mpdaf.obj import Image, WCS, Cube
import numpy as np
from astropy import units as u

data_folder = "/home/carlos/Desktop/pPXF(SDP)/"
```

```python
img = Image(data_folder+'Data/Muse_Image.fits')
img.info()
```

```python
y_px, x_px, Id = np.loadtxt('Full_Spec_Voronoi_OutPut.txt', unpack=True) 
```

```python
img[:]  = np.nan
img.var = img.data.copy() 
img.var[:] = np.nan
```

```python
for i in range(len(y_px)):                               
    img.data.data[int(y_px[i])][int(x_px[i])] = Id[i]
    img.var[int(y_px[i])][int(x_px[i])] = Id[i]
```

```python
img.write(data_folder+'Data/Voronoi/Vorbin_Tag.fits') #Save the new grid
img.info() #Display the information
```
