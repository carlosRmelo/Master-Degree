```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Bin, vel, disp, chi2, dV, dsigma = np.loadtxt('Output_Ppxf_BadAssTemplate.txt', unpack=True) #Reads the table with the adjustment data/LÃª a tabela com os dados do ajuste

vel = vel - vel[0]                                   #vel[0] velocity of central px
vrms = np.sqrt(np.array(vel)**2 + np.array(disp)**2) #Vrms velocity
erms = np.sqrt((np.array(dV)*np.array(vel))**2 + (np.array(dsigma)*np.array(disp))**2)/vrms #error in vrms
```

```python
xnode = np.array([ 0.        , -0.3       , -0.05      ,  0.26666667,  0.3       ,
        0.64      ,  0.35714286, -0.60666667, -1.19285714, -0.46666667,
       -0.2       ,  0.3       , -0.75555556,  0.1       ,  0.76      ,
        1.225     ,  0.5       , -0.13333333,  0.13333333,  1.17857143,
        0.50526316, -0.92142857, -0.18333333])

ynode = np.array([-0.1       , -0.4       , -0.6       , -0.53333333, -0.2       ,
       -0.48      , -1.40714286, -1.22      , -0.22857143, -0.06666667,
        0.1       ,  0.        ,  0.31111111,  0.2       , -0.12      ,
       -0.70625   ,  0.3       ,  0.46666667,  0.53333333,  0.60714286,
        1.17894737,  0.95714286,  1.49166667])

r = np.sqrt(xnode**2 + ynode**2)
```

```python
plt.figure(figsize=(8,6))
plt.errorbar(r,vrms, yerr=erms, fmt=".k", capsize=0)
plt.ylabel("Vrms (km/s)")
plt.xlabel("Circularized radius")
```
