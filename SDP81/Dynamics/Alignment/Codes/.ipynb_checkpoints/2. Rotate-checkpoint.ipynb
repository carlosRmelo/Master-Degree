We want to rotate the images in such a way that the major semi-axis is on the x-axis. This is necessary to be able to compare with the rotated image of ALMA.

To check the overlap in the coutourns, check the folder __Plot contour__.

```python
from mpdaf.obj import Image
data_folder = "/home/carlos/Documents/GitHub/Master-Degree/SDP81/Dynamics/Alignment/Data/"
```

```python
muse = Image(data_folder+"Muse_aligned.fits")                     #Muse image
hst  = Image(data_folder+"hst_12194_01_wfc3_ir_f160w_drz.fits")   #HST image
arcs = Image(data_folder+"arcs_wcs.fits")                         #HST arcs

PA = 11.6   #Position angle [DEG] from HST data, find using find_my_galaxy
```

```python
muse_rot = muse.rotate(90-PA, interp= "spline", reshape=True)
hst_rot  = hst.rotate(90-PA, interp= "spline", reshape=True)
arcs_rot = arcs.rotate(90-PA, interp= "spline", reshape=True)
```

```python
muse_rot.plot(zscale=True, use_wcs=True)
muse_rot.write(data_folder+"Rotate/Muse_aligned_rot.fits")
```

```python
hst_rot.plot(zscale=True, use_wcs=True)
hst_rot.write(data_folder+"Rotate/hst_rot.fits")
```

```python
arcs_rot.plot(use_wcs=True)
arcs_rot.write(data_folder+"Rotate/arcs_rot.fits")
```
