```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
continuum = np.loadtxt("Continuum/Radial_Profile_Continuum.txt")
full      = np.loadtxt("Full/Radial_Profile_Full.txt")
```

```python
plt.figure(figsize=(10,8))

plt.errorbar(continuum[:,0],continuum[:,1], yerr=continuum[:,2], fmt=".k", capsize=0, label="Continuum")
plt.errorbar(full[:,0],full[:,1], yerr=full[:,2], fmt=".r", capsize=0, label="Full")

plt.legend()
plt.xlabel("R [arcsec]")
plt.ylabel("Vrms [km/s]")
```
