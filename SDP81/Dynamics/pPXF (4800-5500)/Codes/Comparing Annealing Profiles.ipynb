```python
import numpy as np
import matplotlib.pyplot as plt
```

## Continuum

```python
continuum_02 = np.loadtxt("Annealing_Bin02/Continuum/Radial_Profile_Continuum.txt")
continuum_03 = np.loadtxt("Annealing_Bin03/Continuum/Radial_Profile_Continuum.txt")
continuum_04 = np.loadtxt("Annealing_Bin04/Continuum/Radial_Profile_Continuum.txt")
```

```python
plt.figure(figsize=(10,8))

plt.errorbar(continuum_02[:,0],continuum_02[:,1], yerr=continuum_02[:,2], fmt="--k", capsize=0, label="Continuum Bin 0.2''")
plt.errorbar(continuum_03[:,0],continuum_03[:,1], yerr=continuum_03[:,2], fmt="--r", capsize=0, label="Continuum Bin 0.3''")
plt.errorbar(continuum_04[:,0],continuum_04[:,1], yerr=continuum_04[:,2], fmt="--b", capsize=0, label="Continuum Bin 0.4''")

plt.legend()
plt.xlabel("R [arcsec]")
plt.ylabel("Vrms [km/s]")
plt.title("Continuum")
```

## Full

```python
full_02 = np.loadtxt("Annealing_Bin02/Full/Radial_Profile_Full.txt")
full_03 = np.loadtxt("Annealing_Bin03/Full/Radial_Profile_Full.txt")
full_04 = np.loadtxt("Annealing_Bin04/Full/Radial_Profile_Full.txt")
```

```python
plt.figure(figsize=(10,8))

plt.errorbar(full_02[:,0],full_02[:,1], yerr=full_02[:,2], fmt="--k", capsize=0, label="Full Bin 0.2''")
plt.errorbar(full_03[:,0],full_03[:,1], yerr=full_03[:,2], fmt="--r", capsize=0, label="Full Bin 0.3''")
plt.errorbar(full_04[:,0],full_04[:,1], yerr=full_04[:,2], fmt="--b", capsize=0, label="Full Bin 0.4''")

plt.legend()
plt.xlabel("R [arcsec]")
plt.ylabel("Vrms [km/s]")
plt.title("Full")
```

## All 

```python
plt.figure(figsize=(10,8))

plt.errorbar(continuum_02[:,0],continuum_02[:,1], yerr=continuum_02[:,2], fmt="--", capsize=0, label="Continuum Bin 0.2''")
plt.errorbar(continuum_03[:,0],continuum_03[:,1], yerr=continuum_03[:,2], fmt="--", capsize=0, label="Continuum Bin 0.3''")
plt.errorbar(continuum_04[:,0],continuum_04[:,1], yerr=continuum_04[:,2], fmt="--", capsize=0, label="Continuum Bin 0.4''")


plt.errorbar(full_02[:,0],full_02[:,1], yerr=full_02[:,2], fmt="--k", capsize=0, label="Full Bin 0.2''")
plt.errorbar(full_03[:,0],full_03[:,1], yerr=full_03[:,2], fmt="--r", capsize=0, label="Full Bin 0.3''")
plt.errorbar(full_04[:,0],full_04[:,1], yerr=full_04[:,2], fmt="--b", capsize=0, label="Full Bin 0.4''")

plt.legend()
plt.xlabel("R [arcsec]")
plt.ylabel("Vrms [km/s]")
plt.title("All")
```
