```python
import numpy as np
import matplotlib.pyplot as plt
import astropy.units as u
from mpdaf.obj import Image, WCS
import astropy.units as u
from mpdaf.obj import deg2sexa, sexa2deg
```

```python
f110w = Image('/home/carlos/Documents/Dados Projeto/SDP.81/Hubble/hst_12194_01_wfc3_ir_f110w_drz.fits')
f160w = Image('/home/carlos/Documents/Dados Projeto/SDP.81/Hubble/hst_12194_01_wfc3_ir_f160w_drz.fits')
```

```python
f110w.info()
f160w.info()
```

```python
#Values by SDSS
RA = '09:03:11.58'
DEC = '+00:39:06.54'
center_sexa = (DEC,RA)
center_deg = sexa2deg(center_sexa)
center_deg
```

```python
corte_f110w = f110w.subimage(center=center_deg, unit_center=u.deg,size=35,unit_size=u.arcsec)
corte_f160w = f160w.subimage(center=center_deg, unit_center=u.deg,size=35,unit_size=u.arcsec)
```

```python
corte_f110w.plot(vmin=0, vmax=1, colorbar='v')
corte_f110w.info()
corte_f110w.write('/home/carlos/Desktop/HST SDP81/Data/Corte_F110w.fits')
```

```python
corte_f160w.plot(vmin=0, vmax=1, colorbar='v',use_wcs=True)
corte_f160w.info()
corte_f160w.write('/home/carlos/Desktop/HST SDP81/Data/Corte_F160w.fits')
```
