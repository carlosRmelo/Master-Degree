```python
from mpdaf.obj import Image
# Import CMasher to register colormaps
import cmasher as cmr
cmap = cmr.pride
import matplotlib.pyplot as plt
```

```python
data_path   = "/home/carlos/Desktop/pPXF(SDP)/Final/Data (SNR25)"
save_folder = "/home/carlos/Documents/GitHub/Master-Degree/Plots for Dissertation/Images/Results"
```

```python
signal = Image(data_path+"/Signal_and_noise.fits", ext=1) #signal data
noise  = Image(data_path+"/Signal_and_noise.fits", ext=2) #noise data

signal.data.data[:] = signal.data.data/noise.data.data    #snr map
```

```python
plt.figure(figsize=(12,8))
cmap = cmr.pride

fig = signal.plot(use_wcs=True, cmap=cmap, colorbar="v")

ax =  plt.subplot(projection=signal.wcs.wcs)
plt.xlim(15,45)
plt.ylim(15,45)

cb = plt.colorbar(fig,  pad=-0.2)
cb.set_label(label="SNR",weight='bold')
#plt.tight_layout()
plt.savefig(save_folder+"/SNR_ppxf.png", fmt="png", dpi=150, bbox_inches='tight')
plt.show()
```

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
vrms_map = Image(data_path+"/Vrms_Map.fits", ext=1)
```

```python
plt.figure(figsize=(12,8))
cmap = cmr.pride

fig = vrms_map.plot(use_wcs=True, cmap=cmap, colorbar="v",vmin=180, vmax=255)

ax =  plt.subplot(projection=vrms_map.wcs.wcs)
plt.xlim(15,45)
plt.ylim(15,45)


cb = plt.colorbar(fig,  pad=-0.2)
cb.set_label(label="km/s",weight='bold')
#plt.tight_layout()
plt.savefig(save_folder+"/Vrms_map.png", fmt="png", dpi=150, bbox_inches='tight')
plt.show()
```
