# Plot dataset for dissertation

To save the images with high quality and better adjusted, press the left mouse button and choose "Save Image As ...".

```python
from mpdaf.obj import Image
import matplotlib.pyplot as plt

# Import CMasher to register colormaps
import cmasher as cmr

save_folder = "/home/carlos/Documents/GitHub/Master-Degree/Plots for Dissertation/Images"

```

```python
f160w = Image("/home/carlos/Desktop/HST Data Reduction/Data/Centered/HST_final.fits")
f160w.info()
f160w.data.data[:] = (f160w.data.data/f160w.data.max())
```

```python
plt.figure(figsize=(12,8))
cmap = cmr.pride

fig = f160w.plot(use_wcs=True, cmap=cmap, colorbar="v", scale="log", vmin=1.35e-2 )

ax =  plt.subplot(projection=f160w.wcs.wcs)

ax.arrow(10, 10, 10, 0,
         head_width=0, head_length=0,
         fc='white', ec='white', width=0.1,
         )
plt.text(11.5, 7, ' 0.9" ',
         color='white', rotation=0,weight='bold'
         )

cb = plt.colorbar(fig,  pad=-0.2)
cb.set_label(label="Flux Normalized by the Maximum ",weight='bold')
#plt.tight_layout()
plt.savefig(save_folder+"/F160W_Original.png", fmt="png", dpi=150)
plt.show()
```
