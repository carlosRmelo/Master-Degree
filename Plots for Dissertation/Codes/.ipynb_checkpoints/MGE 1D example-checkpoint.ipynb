Fit a function using MGE 1d routine.

```python
import numpy as np
import matplotlib.pyplot as plt
from mge_fit_1d import mge_fit_1d 
#To change the labels, we need to edit directly the script mge_fit_1d.py in THIS folder.
```

```python
n = 300  # number of sampled points
x = np.logspace(np.log10(0.01), np.log10(300), n)  # logarithmically spaced radii
y = 100/(1 + x)**4  # The profile must be logarithmically sampled!
plt.clf()
p = mge_fit_1d(x, y, ngauss=16, plot=True,)
```

```python
plt.figure(figsize=(10,8))
p.plot()
```
