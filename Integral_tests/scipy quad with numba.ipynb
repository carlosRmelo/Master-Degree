```python
from scipy.integrate import quad
import  numpy as np
import astropy.units as u
```

```python
def alpha_x(tau, 
            y, x, M, sigma, q):
     
    x_til = x/sigma
    y_til = y/sigma
    eta = np.sqrt(1.0 - q**2)
    eta_sq = 1 - q**2
    
    aux = (M/sigma)*(x_til/(np.sqrt(1 - eta_sq*(tau)**2)))
    exp_arg = (tau**2/2)*(x_til**2 + y_til**2/(1-eta_sq*(tau)**2))
    exp = np.exp(-exp_arg)
    
    sum = np.sum(aux*exp)
    
    function = tau * sum
    
    return function
```

```python
M, sigma, q = np.loadtxt("Input.txt", unpack=True)
grid = np.loadtxt("grid.txt")
grid = (grid*u.arcsec).to(u.rad).value
```

```python
%timeit quad(alpha_x, 0, 1, args=(grid[0][0], grid[0][1], M, sigma, q))
```

```python
from numba import jit

@jit(nopython=True) # Set "nopython" mode for best performance, equivalent to @njit
def jit_alpha_x(tau, 
            y, x, M, sigma, q):
     
    x_til = x/sigma
    y_til = y/sigma
    eta = np.sqrt(1.0 - q**2)
    eta_sq = 1 - q**2
    
    aux = (M/sigma)*(x_til/(np.sqrt(1 - eta_sq*(tau)**2)))
    exp_arg = (tau**2/2)*(x_til**2 + y_til**2/(1-eta_sq*(tau)**2))
    exp = np.exp(-exp_arg)
    
    sum = np.sum(aux*exp)
    
    function = tau * sum
    
    return function
```
