```python
from mpdaf.obj import Cube
from mpdaf.obj import deg2sexa, sexa2deg
import astropy.units as u
```

Abaixo abrimos o cubo original

```python
cube = Cube('/home/carlosmelo/Documents/Projeto/Dados/ESO 325/MUSE data/ADP.2016-09-07T12_23_32.515.fits')
```

Definindo o centro do Cubo e convertendo para outras unidades. O objetivo é utilizar este centro caso seja feito um corte circular.

```python
center_sex = ['-38:10:35.22','13:43:33.350']
center_deg = sexa2deg(center_sex)
center_px = [232.004615992295,275.372461991368] #Direct from Header. Format (y,x)
print(center_deg)
```

Definido um centro, realizamos um corte circular ao redor da região **center**, a unidade é definida por **unit_center**, se a opção for None, então a unidade é px. O raio deste corte é definido por **radius** com unidades definidas por **unit_radius**. **lbda** define o corte em comprimento de onda.

```python
eso_sub = cube.subcube_circle_aperture(center=center_px, radius=3, unit_center=None, unit_radius=u.arcsec,lbda=(4755,5500))
```

Informações sobre o cubo gerado após o corte e display do cubo.

```python
eso_sub.info()
```

```python
cut_imag = eso_sub.sum(axis=0)
cut_imag.plot()
```

Salvando o sub-cubo.

```python
eso_sub.write('eso_corte(3arc).fits')
```

Extração e plot de espectros do sub-cubo.

```python
eso_sub = Cube('eso_corte(3arc).fits')
sp = eso_sub.sum(axis=(1,2))
sp.plot()
sp.write('full_sp.fits')
```

```python
eso_sub = Cube('eso_circular(cut spec 3).fits')
sp = eso_sub.sum(axis=(1,2))
sp.plot()
sp.write('full_sp.fits')
```

Realização de um corte quadrado do cubo original. **size** determina o tamanho do cubo, enquanto que **unit_size** diz as unidades, se a opção for None, então a unidade será px.

```python
corte = cube.subcube(center=center_px, size=6, lbda=(4780,5500), unit_center=None, unit_size=u.arcsec, unit_wave=u.angstrom)
```
