Obj: Perform the binning of spaxels in 0.6''/Realizar o binning nos spaxel em 0.6''

```python
from mpdaf.obj import Cube
from mpdaf.obj import deg2sexa, sexa2deg
from mpdaf.obj import iter_spe
import astropy.units as u
import numpy as np

data_folder = "/home/carlos/Desktop/ESO"
```

Opening the cut datacube, plot it's image and display informations.

Abrindo o cubo de dados já cortado e plotando a imagem.
Exibindo também as informações do cubo.

```python
cube = Cube(data_folder+'/Data/Cut.fits')
cube.info()
image = cube.sum(axis=0)
image.plot(use_wcs=True)
```

```python
for sp,(y,x) in iter_spe(cube, index=True):
        print("Pixel [%d,%d]" % (y, x))
        sp.write(data_folder+'/Data/All_Spec/%dx%d.fits'%(y,x))
        coord_sky = cube.wcs.pix2sky([y,x], unit=u.arcsec)
        y_arcs = coord_sky[0][0]
        x_arcs = coord_sky[0][1] 
        print(y_arcs,x_arcs)
        newrow = [np.int(y), np.int(x), y_arcs, x_arcs]
        if y== 0 and x ==0:
            table = np.array([np.int(y),np.int(x),y_arcs,x_arcs])
        else:
            table = np.vstack([table,newrow])
```

```python
np.savetxt("PixelPosition.txt",table,fmt = "%i  %i %.18e %.18e", header="y[px] \t\t x[px] \t\t y[''] \t\t x['']")
```
