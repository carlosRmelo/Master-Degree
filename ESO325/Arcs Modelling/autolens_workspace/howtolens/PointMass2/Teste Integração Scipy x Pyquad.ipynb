```python
import numpy as np
import scipy.integrate as integrate
import pyquad
from time import perf_counter as clock
from numba import jit
```

```python

def alpha_x(tau, y, x, M, sigma,q):
    
    x_til = x/sigma
    y_til = y/sigma
    eta = np.sqrt(1 - q**2)
    eta_sq = 1 - q**2
    
    aux = (M/sigma)*(x_til/(np.sqrt(1 - eta_sq*(tau)**2)))
    exp_arg = (tau**2/2)*(x_til**2 + y_til**2/(1-eta_sq*(tau)**2))
    exp = np.exp(-exp_arg)
    
    sum = np.sum(aux*exp)
    
    function = tau*sum
    
    return function
```

```python
#Define os parÃ¢metros da MGE

M0 = np.array([3.7, 5, 4.3, 9.8, 3, 5.4, 6])
sigma0 = np.array([1.9, 2.7, 2.3, 13.9, 11.4, 5.7, 7.0])
q0 = np.array([1, 0.74, 0.53, 0.52, 0.56, 0.16, 0.23])
```

```python
grid = np.random.random((10, 2))
grid
```

```python
res = np.zeros(grid.shape[0])

t = clock()
for i in range(res.shape[0]):
    res[i] = integrate.quad(
        alpha_x, 0, 1, args=(grid[i, 0], grid[i, 1], M0, sigma0, q0)
    )[0]

print('Time for %i points: %f'%(res.shape[0], clock()-t))
res
```

```python
res, err = pyquad.quad_grid(alpha_x, 0, 1, grid, (M0,sigma0,q0))
```
