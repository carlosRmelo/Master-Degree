```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import fits
from os import path
%matplotlib notebook

from mpdaf.obj import Image, WCS


from matplotlib.colors import LogNorm



import mgefit
from mgefit.find_galaxy import find_galaxy
from mgefit.mge_fit_1d import mge_fit_1d
from mgefit.sectors_photometry import sectors_photometry
from mgefit.mge_fit_sectors import mge_fit_sectors
from mgefit.mge_print_contours import mge_print_contours
from mgefit.mge_fit_sectors_twist import mge_fit_sectors_twist
from mgefit.sectors_photometry_twist import sectors_photometry_twist
from mgefit.mge_print_contours_twist import mge_print_contours_twist
```

**minlevel:** determina a menor contagem de pixels que será utilizada para realizar a fotometria. Neste caso, pixels com contagem abaixo de 0.2 são ignorados.

**fraction:** determina a fração de pixels (aproximadamente) que serão utilizados no find_galaxy. Se for igual a um, usaremos todos os pixels da imagem.

```python
%matplotlib inline
file = 'PSF.fits'
hdu = fits.open(file)
img = hdu[1].data

skylev = 0.1   # counts/pixel
img -= skylev   # subtract sky
scale = 0.0455  # arcsec/pixel
minlevel = 0.1  # counts/pixel
#ngauss = 12
```

```python
sigmapsf = [0.494, 1.44, 4.71, 13.4]
normpsf = [0.294, 0.559, 0.0813, 0.0657]
```

```python
%matplotlib inline
f = find_galaxy(img, fraction=0.04, plot=1)
```

```python
%matplotlib inline
s = sectors_photometry(img, f.eps, f.theta, f.xpeak, f.ypeak,
                           minlevel=minlevel, plot=1)
```

```python
m = mge_fit_sectors(s.radius, s.angle, s.counts, f.eps,
                        
                        scale=scale, plot=1, bulge_disk=0, linear=0)
```

```python
mge_print_contours(img.clip(minlevel), f.theta, f.xpeak, f.ypeak, m.sol, scale=scale,
                       binning=7, sigmapsf=sigmapsf, normpsf=normpsf, magrange=9)
```

```python
%matplotlib inline
eso = 'Eso325_f814_cut2.fits'
hdu_eso = fits.open(eso)
eso_img = hdu_eso[1].data
eso_img -= 0.1
find_eso = find_galaxy(eso_img, fraction=0.08, plot=1)
```

```python
%matplotlib inline
sectors_eso = sectors_photometry(eso_img, find_eso.eps, find_eso.theta, find_eso.xpeak, find_eso.ypeak,
                           minlevel=0.1, plot=1)
```

```python
%matplotlib inline
m = mge_fit_sectors(sectors_eso.radius, sectors_eso.angle, sectors_eso.counts, find_eso.eps,
                    scale=scale, plot=1, bulge_disk=0, linear=0)
```

```python
%matplotlib inline
mge_print_contours(eso_img.clip(minlevel), find_eso.theta, find_eso.xpeak, find_eso.ypeak, m.sol, scale=scale,
                       )
```
