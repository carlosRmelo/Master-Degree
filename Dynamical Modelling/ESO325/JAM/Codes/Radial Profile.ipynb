```python
from astropy.io import fits
import numpy as np
import matplotlib.pyplot as plt
from time import perf_counter as clock
```

```python
hdu = fits.open('/home/carlos/Desktop/ESO325_HST_DataReduction/Cuts/F475_divided_F814_TimeScaled.fits')
data = hdu[1].data
copy = data
```

```python
plt.imshow(data, vmin=1, vmax=3,origin='lower')
plt.colorbar()
```

```python
plt.imshow(copy,vmin=1,vmax=3)
```

```python
y, x = np.indices((data.shape))
r = np.sqrt((x - 512)**2 + (y - 512)**2)
r = r.astype(np.int)

tbin = np.bincount(r.ravel(), data.ravel())
nr = np.bincount(r.ravel())
radialprofile = tbin / nr
len(radialprofile)
```

```python
plt.plot(radialprofile)
```

```python
nr
```

```python
t = clock()
for i in range(len(radialprofile)):
    count = 0
    for x in range(1025):
        for y in range(1025):
            distance = np.sqrt((x-512)**2 + (y-512)**2)
            if distance >= i and distance < i+1:
                copy[y][x] = radialprofile[i]
                count = count + 1
    print('Raio %d, contagem %d'%(i,count))
print('Time: %f'%(clock()-t))
```

```python
header_new = hdu[1].header
new = fits.PrimaryHDU(copy,header=header_new)
hdul = fits.HDUList([new])
hdul.writeto('/home/carlos/Desktop/ESO325_HST_DataReduction/Cuts/F475w_divided_F814w_circular_timeSca.fits')
```

```python
plt.imshow(copy,vmin=1.4,vmax=1.7, origin='lower')
plt.colorbar()
plt.title('F475w/F814w Time Scalaed and circularized')
plt.savefig('/home/carlos/Desktop/ESO325_HST_DataReduction/Images/F475w_divided_F814w_circular_timeSc.png', fmt='png',dpi=1200)
```
