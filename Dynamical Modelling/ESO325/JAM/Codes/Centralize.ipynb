Obj: After making the cut in the original image of the HST and determining the center by the MGE, this code centers the central pixel of the MGE with the middle of the image.

```python
import numpy as np
import matplotlib.pyplot as plt
import astropy.units as u
from mpdaf.obj import Image, WCS
import astropy.units as u
from mpdaf.obj import deg2sexa, sexa2deg
```

```python
F814W = Image('/home/carlos/Desktop/ESO325_HST_DataReduction/Cuts/Corte_F814w.fits') #Open the image file F814w
F475W = Image('/home/carlos/Desktop/ESO325_HST_DataReduction/Cuts/Corte_F475w.fits') #Open the image file F475w
F814W.info() #Display information about the file
F475W.info() #Display information about the file
```

```python
MGE_CentralPx = (638,594) #Central pixel found by MGE. The format is (y,x)
```

```python
F814W_centerized = F814W.subimage(center=MGE_CentralPx, unit_center=None, size=41., unit_size=u.arcsec)
F475W_centerized = F475W.subimage(center=MGE_CentralPx, unit_center=None, size=41., unit_size=u.arcsec)
    #center: determine the new center
    #unit_center: astropy units of center coordenate. None for px units
    #size: if you disere a new cut, size define the width
    #unit_size: astropy units for the size. None for px units
```

```python
F814W_centerized.info()
F475W_centerized.info()
```

```python
folder_save = '/home/carlos/Desktop/ESO325_HST_DataReduction/Cuts/' #Path for the save folder
F814W_centerized.write(folder_save + 'F814W_centerized.fits') #Save the new centralized images
F475W_centerized.write(folder_save + 'F475W_centerized.fits')
```
