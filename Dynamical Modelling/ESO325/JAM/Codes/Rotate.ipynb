```python
import numpy as np

import matplotlib.pyplot as plt

import astropy.units as u

from mpdaf.obj import Image, WCS
```

```python
original = Image('Corte_F814w.fits')
original.plot(vmin=1,vmax=3,use_wcs=True,colorbar='v')
plt.savefig('F814w_original.png',fmt='png',dpi=1200)
original.info()
```

```python
rotate = original.rotate(22)
rotate.plot(vmin=1,vmax=3,use_wcs=True,colorbar='v')
plt.savefig('F814w_rotate.png',fmt='png',dpi=1200)
rotate.info()
rotate.write('f814w_rotate.fits')
```
