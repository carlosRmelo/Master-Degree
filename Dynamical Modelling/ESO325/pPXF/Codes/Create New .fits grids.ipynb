Obj: Create a new grid with 0.6''/px /Criar um novo grid com 0.6''/px

```python
from mpdaf.obj import Image, WCS, Cube
import numpy as np
from astropy import units as u
```

```python
cube = Cube('/home/carlos/Desktop/ESO325 MUSE-ColletBinning/Data/Cut.fits') #Open the cut datacube/Abre o cubo cortado
image = cube.sum(axis=0) #Create an image/Cria uma imagem
image.plot(use_wcs=True) #Plot the image/Plota a imagem
img_re = image.resample(newdim=(20,20), newstart=None, newstep=(0.6,0.6), unit_step=u.arcsec, flux=True)
                        #Create a new grid with 10x10 pxs and spacial resolution 0.6''. This method automatically corrects WCS.
                        #Cria um novo grid com 10x10 pxs e resolução espacial 0.6''. Esse método automaticamente corrige o WCS.
img_re[:] = 0   #Empty the new grid data/Limpa os dados do novo grid
img_re.var[:] = 0 #Empty the new grid variance/Limpa a variância do novo frid
```

```python
img_re.write('/home/carlos/Desktop/ESO325 MUSE-ColletBinning/Data/New Grid.fits') #Save the new grid
img_re.info() #Display the information
```
